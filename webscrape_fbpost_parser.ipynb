### Prompt Template + LLM

```python
fbpost_template = """\
For the following Facebook post, extract the following information:

topic: What is the topic of the post? Give 1-2 topics in list format. \
If there is a mention of "LuckyMe", include it.

sentiment: What is the sentiment of the post? \
Answer positive, negative or neutral.

summary: Summarize the post in 1-2 sentences.

product_rating: What does the customer say about the quality of the \
product based on his/her post? If there is a mention of product quality, 
give it a rating on a scale of 1 to 5, 5 being the highest." \
"Otherwise, if there is no mention of product quality, based on the sentiment, \
on a scale of 1 to 5, 5 being the highest, what product quality rating can you give?"

Format the output as JSON with the following keys:
topic
sentiment
summary
product_rating

post: {post}
"""

from langchain.prompts import ChatPromptTemplate

prompt_template = ChatPromptTemplate.from_template(fbpost_template)
```

### Prompt Template +LLM + Output Parser

```python
from langchain.output_parsers import ResponseSchema
from langchain.output_parsers import StructuredOutputParser

topic_schema = ResponseSchema(
    name="topic",
    description="What is the main topic of the post. If there are more than 1 topic, \
        do not put in [] instead, separate them with comma. \
        If there is a mention of 'LuckyMe', include it."
)

sentiment_schema = ResponseSchema(
    name="sentiment",
    description="What is the sentiment of the post? \
        Answer positive, negative or neutral."
)

summary_schema = ResponseSchema(
    name="summary",
    description="Summarize the post."
)

product_rating_schema= ResponseSchema(
    name="product_rating",
    description="What does the customer say about the quality of the \
        product based on his/her post? If there is a mention of product quality, \
        give it a rating on a scale of 1 to 5, 5 being the highest. \
        Otherwise, if there is no mention of product quality, based on the sentiment, \
        on a scale of 1 to 5, 5 being the highest, what product quality rating can you give?"
)

response_schemas = [
    topic_schema,
    sentiment_schema,
    summary_schema,
    product_rating_schema
]
```

```python
output_parser = StructuredOutputParser.from_response_schemas(response_schemas)
format_instructions = output_parser.get_format_instructions()
```

```python
og_fbposts = fb['text']
len(og_fbposts)
```

```python
responses_list=[]

for caption in og_fbposts:
    messages = prompt_template.format_messages(
    post=caption, 
    format_instructions=format_instructions
    )
    response = chat(messages)
    output_dict = output_parser.parse(response.content)
    responses_list.append(response.content)
    print(response.content)
```

```python
len(responses_list)
```

```python
# Initialize an empty list to store dictionaries
dict_list = []

# Parse each JSON string and convert it to a dictionary
for json_str in responses_list:
    data_dict = json.loads(json_str)
    dict_list.append(data_dict)

# Create a DataFrame from the list of dictionaries
fbpost_df = pd.DataFrame(dict_list)
```

```python


# Given JSON data (corrected)
data = [
  {
    "topic": [
      "LuckyMe",
      "product quality"
    ],
    "sentiment": "positive",
    "summary": "The post is about LuckyMe and the customer is expressing satisfaction with the product.",
    "product_rating": 5
  },
  {
    "topic": [
      "LuckyMe"
    ],
    "sentiment": "positive",
    "summary": "The post is about chenes chenese.",
    "product_rating": 5
  }
]

# Create a DataFrame from the JSON data
df = pd.DataFrame(data)

# Create a list to store unique topic values
unique_topics = []

# Iterate through the "topic" column to find unique topics
for topics_list in df['topic']:
    for topic in topics_list:
        unique_topics.append(topic.lower())

# Remove duplicates to get unique topics
unique_topics = list(set(unique_topics))

# Create a new DataFrame with one-hot encoding for each unique topic
one_hot_df = pd.DataFrame(0, columns=unique_topics, index=df.index)

# Iterate through the original DataFrame and set one-hot encoding values
for i, topics_list in enumerate(df['topic']):
    for topic in topics_list:
        one_hot_df.loc[i, topic.lower()] = 1

# Concatenate the one-hot encoded DataFrame with the original DataFrame
result_df = pd.concat([df.drop(columns='topic'), one_hot_df], axis=1)

# Print the result DataFrame
result_df
```
